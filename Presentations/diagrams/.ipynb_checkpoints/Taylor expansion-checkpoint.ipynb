```python
import numpy as np
import tensorflow as tf
from functools import partial
import matplotlib.pyplot as plt
from scipy import stats

sess = tf.Session()

%matplotlib inline
```

```python
a = 2.6
b = 5.0

factorial = lambda n: tf.exp(tf.lgamma(n + 1))

def func(x, a=a, b=b):
    def beta(a,b):
        return tf.exp(tf.math.lgamma(a))*tf.exp(tf.math.lgamma(b)) / tf.exp(tf.math.lgamma(a+b))

    return (x**(a-1) * (1-x)**(b-1)) / beta(a,b)

def log_func(x, a=a, b=b):
    return tf.log(func(x, a=a, b=b))

def mode(a=a, b=b):
    return tf.constant((a-1) / (a + b - 2), dtype=tf.float32)

def taylor_series(function, x, point, truncate):
    eval_taylor = function(point)
    grad = function(point)
    for i in range(1, truncate+1):
        i = float(i)
        grad = tf.gradients(grad, point)
        eval_taylor += (grad / factorial(i))*(x - point)**i
    return eval_taylor, grad
```

```python
x = np.linspace(0.0,1.0,100).astype(np.float64)
f = func
log_f = func
m = mode
plt.plot(x, sess.run(f(x)), color='C0', label='$h(f)$')
plt.axvline(sess.run(m()), color='r', linestyle='--', label='Mode $\hat{f}$')
plt.legend()
```

```python
x_log = np.linspace(0.0, 0.5, 100)

for n in range(1, 5):
    taylor, _ = sess.run(taylor_series(
        function=log_f,
        x=x_log, 
        point=m(),
        truncate=n))

    plt.plot(x_log, taylor, 'C{}'.format(n), label='taylor $n$={}'.format(n) + ' at $\hat{f}$')

plt.plot(x_log, sess.run(log_f(x_log)), 'C0', label='$log h(f)$')
plt.axvline(sess.run(m()), color='r', linestyle='--', label='Mode $\hat{f}$')
plt.legend()
```

```python
taylor, hessian = sess.run(taylor_series(
        function=log_f,
        x=x_log, 
        point=m(),
        truncate=2))
plt.plot(x_log, taylor, 'C{}'.format(2), label='taylor $n$=2 at $\hat{f}$')

plt.plot(x_log, sess.run(log_f(x_log)), 'C0', label='log h(f)')
plt.axvline(sess.run(m()), color='r', linestyle='--', label='Mode')
plt.legend()
```

Now we move back into probability space, and use a Gaussian approximation with this hessian

```python
W = 1.0/(-hessian[0])
```

```python
plt.plot(x,
         stats.norm.pdf(x=x, 
                        loc = sess.run(m()),
                        scale = np.sqrt(W)),
         'C2', label='$N(f|\hat{f}, \frac{d^{2} log h(\hat{f})}{\hat{f}^{2}})$'
        )
plt.plot(x, sess.run(f(x)), 'C0', label='h(f)')
plt.axvline(sess.run(m()), color='r', linestyle='--', label='Mode $\hat{f}$')
plt.legend()
```
